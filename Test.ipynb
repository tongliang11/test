```python
def averagea(value):
    return sum(v for v in value if len(value) > 0)/len(value)
```

```python
averagea([1,2,3,4])
```

```python
def total(scale, *arg):
    return sum(a*scale for a in arg)
```

```python
total(0.5,1,1,2,3,4,3,4,5)
```

```python
def firstn(value):
    for v in value:
        
        if v < 0:
            return v
        
```

```python
firstn([2,3,1])
```

```python
firstn([])
```

```python
def running(values):
    '''
    This is a function I'm gonna document.
    '''
    if len(values)==0:
        return None
    
    s=[0]*len(values);
    
    for i in range(len(values)+1):
        s[i-1]=sum(v for v in values[:i])
        
    print(s)
        
```

```python
running([1,5])
```
